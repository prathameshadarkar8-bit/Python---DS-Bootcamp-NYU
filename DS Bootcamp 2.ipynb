```python
#Numpy 1
import numpy as np


A = np.array([[5, 6, 7],
              [8, 9, 10]])

B = np.array([[11, 12, 13],
              [14, 15, 16]])

print("Array A:")
print(A)

print("\nArray B:")
print(B)

vertical_stack = np.vstack((A, B))
print("\nVertical Stack:")
print(vertical_stack)

horizontal_stack = np.hstack((A, B))
print("\nHorizontal Stack:")
print(horizontal_stack)
```

```python
#Numpy 2
common = np.intersect1d(A, B)
print("\nCommon elements:", common)
```

```python
#Numpy 3
indices = np.where((A >= 5) & (A <= 10))[0]  # [0] to get the array from tuple
result = A[indices]
print("Using np.where():", result)
```

```python
#Numpy 4
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

filtered_iris = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)]

print("\nFiltered iris_2d:")
print(f"Shape: {filtered_iris.shape}")
print("\nFirst 10 filtered rows:")
print(filtered_iris[:10])

print(f"\nTotal rows matching criteria: {filtered_iris.shape[0]} out of {iris_2d.shape[0]}")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

print("Missing values before:")
print(df[['Min.Price', 'Max.Price']].isnull().sum())

df['Min.Price'] = df['Min.Price'].fillna(df['Min.Price'].mean())
df['Max.Price'] = df['Max.Price'].fillna(df['Max.Price'].mean())

print("\nMissing values after:")
print(df[['Min.Price', 'Max.Price']].isnull().sum())
print(df[['Min.Price', 'Max.Price']])
```
