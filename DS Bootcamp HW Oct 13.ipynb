```python
#Q1 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 100

# Read the CSV file 
df = pd.read_csv('/Users/prathameshadarkar/Downloads/BB.csv')

# Convert hour_beginning to datetime
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])

# Extract day of week 
df['day_of_week'] = df['hour_beginning'].dt.dayofweek
df['day_name'] = df['hour_beginning'].dt.day_name()
df['hour'] = df['hour_beginning'].dt.hour
df['month'] = df['hour_beginning'].dt.month

# Filter for only weekdays 
weekday_df = df[df['day_of_week'] < 5].copy()

# Convert Pedestrians column to numeric
weekday_df['Pedestrians'] = pd.to_numeric(weekday_df['Pedestrians'], errors='coerce')

# Remove any NaN values from key columns
weekday_df = weekday_df.dropna(subset=['Pedestrians', 'temperature', 'precipitation'])

# Create a pairplot 
print("Creating pairplot...")
pairplot_data = weekday_df[['Pedestrians', 'temperature', 'precipitation', 'hour', 'month']].copy()

# Create pairplot
pairplot = sns.pairplot(pairplot_data, 
                        vars=['Pedestrians', 'temperature', 'precipitation', 'hour', 'month'],
                        plot_kws={'alpha': 0.5, 'size': 0.1},
                        diag_kind='hist',
                        corner=False)
pairplot.fig.suptitle('Brooklyn Bridge Weekday Data - Pairplot Analysis', 
                      y=1.02, fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nPairplot created successfully!\n")

# Group by day name and calculate average pedestrian counts
daily_avg = weekday_df.groupby('day_name')['Pedestrians'].mean().reset_index()

# Ensure proper ordering of days (Monday to Friday)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
daily_avg['day_name'] = pd.Categorical(daily_avg['day_name'], categories=day_order, ordered=True)
daily_avg = daily_avg.sort_values('day_name')

# Create the line graph
plt.figure(figsize=(12, 6))
plt.plot(daily_avg['day_name'], daily_avg['Pedestrians'], 
         marker='o', linewidth=2.5, markersize=10, color='#2563eb')

# Customize the plot
plt.title('Brooklyn Bridge - Average Pedestrian Counts by Weekday', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Day of Week', fontsize=12, fontweight='bold')
plt.ylabel('Average Pedestrian Count', fontsize=12, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(rotation=45, ha='right')

# Add value labels on each point
for i, row in daily_avg.iterrows():
    plt.text(row['day_name'], row['Pedestrians'], 
             f"{row['Pedestrians']:.0f}", 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

# Print summary statistics
print("=" * 60)
print("WEEKDAY PEDESTRIAN TRAFFIC SUMMARY")
print("=" * 60)
print(f"\nTotal weekday observations: {len(weekday_df):,}")
print(f"\nAverage pedestrian counts by day:\n")
for _, row in daily_avg.iterrows():
    print(f"  {row['day_name']:12s}: {row['Pedestrians']:8.0f} pedestrians")

print("\n" + "=" * 60)
print(f"Highest traffic day: {daily_avg.loc[daily_avg['Pedestrians'].idxmax(), 'day_name']}")
print(f"Lowest traffic day:  {daily_avg.loc[daily_avg['Pedestrians'].idxmin(), 'day_name']}")
print("=" * 60)

#Q2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style 
sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 100

# Read the CSV file 
df = pd.read_csv('/Users/prathameshadarkar/Downloads/BB.csv')

# Convert hour_beginning to datetime
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])

# Filter data for 2019
df_2019 = df[df['hour_beginning'].dt.year == 2019].copy()

print("=" * 80)
print("BROOKLYN BRIDGE PEDESTRIAN ANALYSIS - YEAR 2019")
print("=" * 80)
print(f"\nTotal records for 2019: {len(df_2019):,}")

# Convert Pedestrians column to numeric
df_2019['Pedestrians'] = pd.to_numeric(df_2019['Pedestrians'], errors='coerce')
df_2019['temperature'] = pd.to_numeric(df_2019['temperature'], errors='coerce')
df_2019['precipitation'] = pd.to_numeric(df_2019['precipitation'], errors='coerce')

# Remove NaN values
df_2019_clean = df_2019.dropna(subset=['Pedestrians', 'temperature', 'precipitation']).copy()

print(f"Records after cleaning: {len(df_2019_clean):,}\n")


# 1. SORT AND ANALYZE BY WEATHER SUMMARY
print("=" * 80)
print("PEDESTRIAN COUNTS BY WEATHER SUMMARY")
print("=" * 80)

# Sort by weather summary and calculate statistics
weather_stats = df_2019_clean.groupby('weather_summary').agg({
    'Pedestrians': ['count', 'mean', 'median', 'std', 'min', 'max']
}).round(2)

weather_stats.columns = ['Count', 'Mean', 'Median', 'Std Dev', 'Min', 'Max']
weather_stats = weather_stats.sort_values('Mean', ascending=False)

print("\n", weather_stats)
print("\n")


# 2. CORRELATION MATRIX
print("=" * 80)
print("CORRELATION ANALYSIS")
print("=" * 80)

# Select numeric columns for correlation analysis
correlation_cols = ['Pedestrians', 'temperature', 'precipitation']
correlation_data = df_2019_clean[correlation_cols].copy()

# Calculate correlation matrix
correlation_matrix = correlation_data.corr()

print("\nCorrelation Matrix:")
print(correlation_matrix.round(3))
print("\n")


# 3. VISUALIZATIONS
# Create figure with multiple subplots
fig = plt.figure(figsize=(16, 12))

# Plot 1: Correlation Matrix Heatmap
plt.subplot(2, 3, 1)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1, fmt='.3f')
plt.title('Correlation Matrix - 2019', fontsize=12, fontweight='bold')

# Plot 2: Pedestrian Count by Weather Summary (Box Plot)
plt.subplot(2, 3, 2)
weather_order = weather_stats.index.tolist()
sns.boxplot(data=df_2019_clean, y='weather_summary', x='Pedestrians', 
            order=weather_order, palette='Set2')
plt.title('Pedestrian Distribution by Weather', fontsize=12, fontweight='bold')
plt.xlabel('Pedestrian Count')
plt.ylabel('Weather Summary')

# Plot 3: Average Pedestrian Count by Weather
plt.subplot(2, 3, 3)
weather_means = df_2019_clean.groupby('weather_summary')['Pedestrians'].mean().sort_values(ascending=False)
bars = plt.barh(range(len(weather_means)), weather_means.values, color='steelblue')
plt.yticks(range(len(weather_means)), weather_means.index)
plt.xlabel('Average Pedestrian Count')
plt.title('Average Pedestrians by Weather', fontsize=12, fontweight='bold')
plt.grid(axis='x', alpha=0.3)

# Add value labels on bars
for i, (idx, val) in enumerate(weather_means.items()):
    plt.text(val, i, f' {val:.0f}', va='center', fontsize=9)

# Plot 4: Temperature vs Pedestrians Scatter
plt.subplot(2, 3, 4)
plt.scatter(df_2019_clean['temperature'], df_2019_clean['Pedestrians'], 
            alpha=0.3, s=10, color='coral')
plt.xlabel('Temperature (F)')
plt.ylabel('Pedestrian Count')
plt.title('Temperature vs Pedestrian Count', fontsize=12, fontweight='bold')
plt.grid(alpha=0.3)

# Add trend line
z = np.polyfit(df_2019_clean['temperature'].dropna(), 
               df_2019_clean.loc[df_2019_clean['temperature'].notna(), 'Pedestrians'], 1)
p = np.poly1d(z)
plt.plot(df_2019_clean['temperature'].sort_values(), 
         p(df_2019_clean['temperature'].sort_values()), 
         "r--", linewidth=2, alpha=0.8, label=f'Trend: y={z[0]:.2f}x+{z[1]:.2f}')
plt.legend()

# Plot 5: Precipitation vs Pedestrians Scatter
plt.subplot(2, 3, 5)
plt.scatter(df_2019_clean['precipitation'], df_2019_clean['Pedestrians'], 
            alpha=0.3, s=10, color='skyblue')
plt.xlabel('Precipitation (inches)')
plt.ylabel('Pedestrian Count')
plt.title('Precipitation vs Pedestrian Count', fontsize=12, fontweight='bold')
plt.grid(alpha=0.3)

# Plot 6: Monthly Pedestrian Trends
plt.subplot(2, 3, 6)
df_2019_clean['month'] = df_2019_clean['hour_beginning'].dt.month
monthly_avg = df_2019_clean.groupby('month')['Pedestrians'].mean()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.plot(monthly_avg.index, monthly_avg.values, marker='o', 
         linewidth=2.5, markersize=8, color='green')
plt.xticks(range(1, 13), months, rotation=45)
plt.xlabel('Month')
plt.ylabel('Average Pedestrian Count')
plt.title('Monthly Pedestrian Trends - 2019', fontsize=12, fontweight='bold')
plt.grid(alpha=0.3)

plt.suptitle('Brooklyn Bridge Pedestrian Analysis - 2019', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()

# 4. KEY INSIGHTS
print("=" * 80)
print("KEY INSIGHTS")
print("=" * 80)

# Find correlations
temp_corr = correlation_matrix.loc['Pedestrians', 'temperature']
precip_corr = correlation_matrix.loc['Pedestrians', 'precipitation']

print(f"\n1. Temperature Correlation: {temp_corr:.3f}")
if temp_corr > 0.3:
    print("    Strong positive correlation: Warmer weather attracts more pedestrians")
elif temp_corr > 0:
    print("    Moderate positive correlation: Temperature has some influence")
else:
    print("    Negative or no correlation")

print(f"\n2. Precipitation Correlation: {precip_corr:.3f}")
if precip_corr < -0.3:
    print("    Strong negative correlation: Rain significantly reduces pedestrian traffic")
elif precip_corr < 0:
    print("    Moderate negative correlation: Rain has some deterrent effect")
else:
    print("    No negative correlation")

print(f"\n3. Best Weather for Pedestrians: {weather_means.index[0]}")
print(f"    Average count: {weather_means.values[0]:.0f} pedestrians")

print(f"\n4. Worst Weather for Pedestrians: {weather_means.index[-1]}")
print(f"    Average count: {weather_means.values[-1]:.0f} pedestrians")

print("\n" + "=" * 80)

#Q3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 100

# Read the CSV file (use full path for Mac)
df = pd.read_csv('/Users/prathameshadarkar/Downloads/BB.csv')

# Convert hour_beginning to datetime
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])

# Convert Pedestrians column to numeric
df['Pedestrians'] = pd.to_numeric(df['Pedestrians'], errors='coerce')

# Remove NaN values
df_clean = df.dropna(subset=['Pedestrians']).copy()


# CUSTOM FUNCTION TO CATEGORIZE TIME OF DAY

def categorize_time_of_day(hour):
    """
    Categorize hour into time of day periods.
    
    Parameters:
    hour (int): Hour of the day (0-23)
    
    Returns:
    str: Time category ('Morning', 'Afternoon', 'Evening', 'Night')
    
    Categories:
    - Morning: 6 AM - 11 AM (6-11)
    - Afternoon: 12 PM - 5 PM (12-17)
    - Evening: 6 PM - 9 PM (18-21)
    - Night: 10 PM - 5 AM (22-5)
    """
    if 6 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 17:
        return 'Afternoon'
    elif 18 <= hour <= 21:
        return 'Evening'
    else:  # 22-23 and 0-5
        return 'Night'

# Extract hour from datetime
df_clean['hour'] = df_clean['hour_beginning'].dt.hour

# Apply custom function to create new column
df_clean['time_of_day'] = df_clean['hour'].apply(categorize_time_of_day)

print("=" * 80)
print("TIME OF DAY PEDESTRIAN ACTIVITY ANALYSIS")
print("=" * 80)
print(f"\nTotal records analyzed: {len(df_clean):,}")
print(f"\nNew column 'time_of_day' created successfully!")
print("\nTime of Day Categories:")
print("  - Morning:   6 AM - 11 AM")
print("  - Afternoon: 12 PM - 5 PM")
print("  - Evening:   6 PM - 9 PM")
print("  - Night:     10 PM - 5 AM")

# ANALYZE PEDESTRIAN ACTIVITY BY TIME OF DAY

print("\n" + "=" * 80)
print("PEDESTRIAN ACTIVITY STATISTICS BY TIME OF DAY")
print("=" * 80)

# Calculate statistics for each time period
time_stats = df_clean.groupby('time_of_day').agg({
    'Pedestrians': ['count', 'mean', 'median', 'std', 'min', 'max', 'sum']
}).round(2)

time_stats.columns = ['Count', 'Mean', 'Median', 'Std Dev', 'Min', 'Max', 'Total']

# Order by custom time sequence
time_order = ['Morning', 'Afternoon', 'Evening', 'Night']
time_stats = time_stats.reindex(time_order)

print("\n", time_stats)

# Calculate percentage of total pedestrians
total_pedestrians = df_clean['Pedestrians'].sum()
time_stats['Percentage'] = (time_stats['Total'] / total_pedestrians * 100).round(2)

print("\n\nPedestrian Distribution:")
for period in time_order:
    pct = time_stats.loc[period, 'Percentage']
    total = time_stats.loc[period, 'Total']
    print(f"  {period:12s}: {pct:6.2f}% ({total:,.0f} pedestrians)")

# VISUALIZATIONS

fig = plt.figure(figsize=(16, 10))

# Plot 1: Average Pedestrians by Time of Day (Bar Chart)
plt.subplot(2, 3, 1)
avg_by_time = df_clean.groupby('time_of_day')['Pedestrians'].mean().reindex(time_order)
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
bars = plt.bar(time_order, avg_by_time.values, color=colors, edgecolor='black', linewidth=1.5)
plt.ylabel('Average Pedestrian Count', fontweight='bold')
plt.title('Average Pedestrian Count by Time of Day', fontsize=12, fontweight='bold', pad=10)
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}', ha='center', va='bottom', fontweight='bold')

# Plot 2: Distribution Box Plot
plt.subplot(2, 3, 2)
sns.boxplot(data=df_clean, x='time_of_day', y='Pedestrians', 
            order=time_order, palette=colors)
plt.xlabel('Time of Day', fontweight='bold')
plt.ylabel('Pedestrian Count', fontweight='bold')
plt.title('Pedestrian Count Distribution', fontsize=12, fontweight='bold', pad=10)
plt.xticks(rotation=0)

# Plot 3: Pie Chart - Percentage Distribution
plt.subplot(2, 3, 3)
sizes = [time_stats.loc[period, 'Percentage'] for period in time_order]
plt.pie(sizes, labels=time_order, autopct='%1.1f%%', startangle=90, 
        colors=colors, textprops={'fontsize': 10, 'fontweight': 'bold'})
plt.title('Pedestrian Distribution by Time Period', fontsize=12, fontweight='bold', pad=10)

# Plot 4: Hourly Pattern Throughout the Day
plt.subplot(2, 3, 4)
hourly_avg = df_clean.groupby('hour')['Pedestrians'].mean()
plt.plot(hourly_avg.index, hourly_avg.values, marker='o', linewidth=2.5, 
         markersize=6, color='#2C3E50')

# Add colored background for each time period
plt.axvspan(6, 11, alpha=0.2, color=colors[0], label='Morning')
plt.axvspan(12, 17, alpha=0.2, color=colors[1], label='Afternoon')
plt.axvspan(18, 21, alpha=0.2, color=colors[2], label='Evening')
plt.axvspan(22, 24, alpha=0.2, color=colors[3])
plt.axvspan(0, 5, alpha=0.2, color=colors[3], label='Night')

plt.xlabel('Hour of Day', fontweight='bold')
plt.ylabel('Average Pedestrian Count', fontweight='bold')
plt.title('Hourly Pedestrian Pattern', fontsize=12, fontweight='bold', pad=10)
plt.xticks(range(0, 24, 2))
plt.legend(loc='upper left')
plt.grid(alpha=0.3)

# Plot 5: Violin Plot
plt.subplot(2, 3, 5)
sns.violinplot(data=df_clean, x='time_of_day', y='Pedestrians', 
               order=time_order, palette=colors)
plt.xlabel('Time of Day', fontweight='bold')
plt.ylabel('Pedestrian Count', fontweight='bold')
plt.title('Pedestrian Count Density Distribution', fontsize=12, fontweight='bold', pad=10)

# Plot 6: Total Pedestrians by Time Period
plt.subplot(2, 3, 6)
total_by_time = df_clean.groupby('time_of_day')['Pedestrians'].sum().reindex(time_order)
bars = plt.bar(time_order, total_by_time.values, color=colors, edgecolor='black', linewidth=1.5)
plt.ylabel('Total Pedestrian Count', fontweight='bold')
plt.title('Total Pedestrian Traffic by Time Period', fontsize=12, fontweight='bold', pad=10)
plt.grid(axis='y', alpha=0.3)

# Add value labels
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:,.0f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.suptitle('Brooklyn Bridge - Time of Day Pedestrian Activity Analysis', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()


# KEY INSIGHTS

print("\n" + "=" * 80)
print("KEY INSIGHTS")
print("=" * 80)

# Find busiest and quietest periods
busiest_period = avg_by_time.idxmax()
quietest_period = avg_by_time.idxmin()
busiest_count = avg_by_time.max()
quietest_count = avg_by_time.min()

print(f"\n1. Busiest Time Period: {busiest_period}")
print(f"    Average: {busiest_count:.0f} pedestrians")

print(f"\n2. Quietest Time Period: {quietest_period}")
print(f"    Average: {quietest_count:.0f} pedestrians")

print(f"\n3. Peak Activity Difference: {busiest_count - quietest_count:.0f} pedestrians")
print(f"    {((busiest_count / quietest_count - 1) * 100):.1f}% more pedestrians during {busiest_period}")

# Find peak hour
peak_hour = hourly_avg.idxmax()
peak_hour_count = hourly_avg.max()
print(f"\n4. Peak Hour: {peak_hour}:00")
print(f"    Average: {peak_hour_count:.0f} pedestrians")

# Activity pattern description
print("\n5. Activity Pattern:")
sorted_periods = avg_by_time.sort_values(ascending=False)
for i, (period, count) in enumerate(sorted_periods.items(), 1):
    print(f"   {i}. {period}: {count:.0f} avg pedestrians")

print("\n" + "=" * 80)

# Display first few rows with new column
print("\nSample Data with New 'time_of_day' Column:")
print(df_clean[['hour_beginning', 'hour', 'time_of_day', 'Pedestrians']].head(10))
print("\n" + "=" * 80)





```
